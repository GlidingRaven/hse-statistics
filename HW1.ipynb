```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# BMI Dataset from https://www.kaggle.com/datasets/yasserh/bmidataset

# take only Height
df = pd.read_csv('bmi.csv')[['Gender','Height']]
df.shape
```

```python
unique_height = sorted(df['Height'].unique())
print(len(unique_height), 'unuque values from', min(unique_height), 'to', max(unique_height), 'cm')
```

### Distribution function

```python
stat = []

for el in unique_height:
    stat.append([el, df[df['Height'] < el].shape[0] / df.shape[0]])
    
stat = pd.DataFrame(stat, columns = ['Height','prob'])
plt.plot(stat['Height'], stat['prob'])
plt.show()
```

In this case, I expected the curve to be S-shaped. Let's look at the histogram

```python
df['Height'].hist(bins=30)
```

Histogram more looks like an uniform distribution, thats why we got uniformly increasing distribution function. This is a characteristic (особенность по русски) of the original dataset
